```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Load the dataset
data = pd.read_csv(r"C:\Users\lenovo\Downloads\superstore.csv")
```

```python
# Convert 'Order.Date' to datetime and aggregate Sales by date
data['Order.Date'] = pd.to_datetime(data['Order.Date'], errors='coerce')
daily_sales = data.groupby('Order.Date')['Sales'].sum().dropna()
```

```python
# Define a custom colormap
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]
custom_cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

# Plot using the custom colormap
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot of daily sales, coloring points by sales values
sc = ax.scatter(daily_sales.index, daily_sales.values, c=daily_sales.values, cmap=custom_cmap)
ax.set_xlabel('Order Date')
ax.set_ylabel('Sales')
ax.set_title('Daily Sales with Custom Colormap')

# Add colorbar with correct axis
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Sales Value')

plt.show()
```
